```python
import numpy as np
import pandas as pd
import sys, os, time
main_folder = './../'
if main_folder not in sys.path:
    sys.path.append(main_folder)
from evolution import ClusteredDifferentialEvolution
from glob       import glob
from tqdm       import tqdm

import dirs
from utils      import get_solution, ProgressBar, write_log
import matplotlib.pyplot as plt 

progbar = ProgressBar(elapsed_time=True)
```

### Parameters

```python
# Competition Parameters
funcId      = 6
dim         = 10
numRuns     = 10
targetError = 1e-8

# Algorithm Parameters
pop_size = 30
percMaxEvals = 1
maxEvals = 'auto'
maxGenerations = None
fitness_clusters = 5
cluster_rate = .01
crossover = 'binonial'
mutation = 'best'
n_diff = 1
lambda_mutation = 1
opposition = False
input_data_filepath = './../input_data'
active = False
```

```python
maxEvals = maxEvals if percMaxEvals is None else percMaxEvals*10000*dim
run_tag = '[' + '_'.join([str(funcId), 
                          str(dim), 
                          str(numRuns),
                          str(pop_size), 
                          str(percMaxEvals),                                                     
                          '{}'.format(lambda_mutation), 
                          str(opposition),
                          'variable'
                         ])  + ']'
run_tag
```

```python
alg = ClusteredDifferentialEvolution(dim=dim, func_id=funcId, cluster_rate=cluster_rate, pop_size=pop_size, crossover=crossover, 
        mutation=mutation, lambda_mutation=lambda_mutation)

output_filepath = os.path.join(dirs.results, str(alg))
if not os.path.exists(output_filepath):
    os.makedirs(output_filepath)
output_filepath = output_filepath + '/' + run_tag

# print ('Writing log at ', output_filepath + 'log.tsv')
# write_log(filepath=output_filepath + 'log.tsv', 
#           mode="w+", 
#           text='[{}]\trun\telapsedTime\tsuccessRate\n'.format(time.strftime("%Y-%m-%d %H:%M:%S", time.gmtime())))  
```

```python
arr_cluster_rate = [0.001, 0.005, 0.01, 0.05, 0.1]
arr_fitness_clusters = [3, 5, 8, 10]
```

```python
hist = pd.DataFrame()
arr_runs = range(numRuns)
solution = get_solution(funcId, dim, input_data_filepath=input_data_filepath)
fig, ax = plt.subplots(figsize=(10,10))
for cluster_rate in arr_cluster_rate:
    for fitness_clusters in arr_fitness_clusters:
        print ('Starting fitness cluster {} on cluster rate {}'.format(fitness_clusters, cluster_rate))
        successRate = 0
        max_plots = 6
        for idx_run in arr_runs:
            progbar.update_progress(idx_run/float(len(arr_runs)))

            # Differential Evolution
            alg = ClusteredDifferentialEvolution(dim=dim, func_id=funcId, cluster_rate=cluster_rate, pop_size=pop_size, crossover=crossover, 
                mutation=mutation, lambda_mutation=lambda_mutation, fitness_clusters=fitness_clusters)

            errorHist, fitnessHist = alg.optimize(target=solution, max_f_evals=maxEvals, target_error=targetError, verbose=True)        
            errorHist['fitness_clusters'] = fitness_clusters
            errorHist['cluster_rate'] = cluster_rate
            bestError = errorHist.iloc[-1,:].min()
            errorHist["run"] = np.ones(errorHist.shape[0], dtype=int)*idx_run
            hist = pd.concat([hist, errorHist], ignore_index=False)    
            if bestError <= targetError:
                print ("Target error achieved with error {}".format(bestError))
                successRate += 1

        #     write_log(filepath=output_filepath + 'log.tsv', 
        #               text='[{}]\t{}/{}\t{}\t{}\n'.format(time.strftime("%Y-%m-%d %H:%M:%S", time.gmtime()), 
        #                                               idx_run+1,
        #                                               len(arr_runs),
        #                                               progbar.get_elapsed_time(),
        #                                               successRate)
        #              )

            if (idx_run < max_plots):
                arr = errorHist.drop(['run', 'fitness_clusters', 'cluster_rate'], axis=1).min(axis=1)    
                ax.scatter(np.arange(alg.generations), 
                           [None if x == pop_size else arr[i] for i, x in enumerate(alg.fitnessEvalsHist[:-1])],
                           marker='.'
                          )
                ax.plot(arr, label='run ' + str(idx_run) + ' Fit' + str(fitness_clusters) + ' ClR' + str(cluster_rate))
    
ax.set_yscale('log')
ax.grid(True)
ax.set_title('CDE Error Evolution (max {} plot)'.format(max_plots))
ax.set_xlabel('Generation')
ax.set_ylabel('Fitness Error')
ax.legend()
# fig.savefig(output_filepath + 'fitness_evolution.jpeg', bbox_inches='tight')

hist.index.name = 'generation'
hist.to_csv(output_filepath + 'hist.tsv', sep='\t')
progbar.update_progress(1)
```

```python
arr_mean = np.zeros([len(arr_fitness_clusters), len(arr_cluster_rate)])
arr_std = np.zeros([len(arr_fitness_clusters), len(arr_cluster_rate)])
for i_cluster_rate, cr in enumerate(arr_cluster_rate):
    for i_fitness_clusters, fc in enumerate(arr_fitness_clusters):        
        hist_temp = hist[hist['cluster_rate'] == cr]
        hist_temp = hist_temp[hist_temp['fitness_clusters'] == fc]
        hist_temp = hist_temp.groupby(['run']).min().reset_index(drop = False)[[str(x) for x in np.arange(pop_size)]].min(axis=1)
        arr_mean[i_fitness_clusters][i_cluster_rate] = hist_temp.mean()
        arr_std[i_fitness_clusters][i_cluster_rate] = hist_temp.std()
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(8,6))
im = ax.pcolormesh(arr_mean)
ax.set_xticks([x + 0.5 for x in np.arange(len(arr_cluster_rate))])
ax.set_xticklabels(arr_cluster_rate)
ax.set_yticks([x + 0.5 for x in np.arange(len(arr_fitness_clusters))])
ax.set_yticklabels(arr_fitness_clusters)
ax.set_xlabel('Cluster Rate') 
ax.set_ylabel('Fitness Clusters')
ax.set_title('Mean Minimum Fitness Error for {} runs'.format(numRuns))

fig.colorbar(im, ax=ax)
fig.savefig('./../results/General/cde_parameter_sweep.jpeg', bbox_inches = 'tight')
```
